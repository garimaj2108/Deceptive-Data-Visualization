```python
import glob, os
import pandas as pd
```

```python
path =r'/Users/garimajain/Desktop/Deceptive_Visualization' # use your path
all_files = glob.glob(os.path.join(path, "*.csv"))     # advisable to use os.path.join as this makes concatenation OS independent

df_from_each_file = (pd.read_csv(f) for f in all_files)
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
concatenated_df
```

```python
del concatenated_df['Operations']
```

```python
del concatenated_df['Address']
```

```python
del concatenated_df['City Or County']
```

```python
Total_Killed = concatenated_df['# Killed'].sum()
```

```python
concatenated_df.loc['Total_Killed'] = pd.Series(concatenated_df['# Killed'].sum(), index = ['# Killed'])
```

```python
Total_Injured = concatenated_df['# Injured'].sum()
```

```python
concatenated_df.loc['Total_Injured'] = pd.Series(concatenated_df['# Injured'].sum(), index = ['# Injured'])
```
